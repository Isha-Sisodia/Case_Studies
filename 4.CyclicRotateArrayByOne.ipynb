```python
a = [1,2,3,4,5]
last = a[-1]
for i in range(len(a)-1, 0,-1):
  a[i] = a[i-1]
a[0] = last
print(a)
```

```python
# Submitted code
class Solution:
    def rotate(self, arr):
        last = arr[-1]
        for i in range(len(arr)-1,0,-1):
            arr[i] =arr[i-1]
        arr[0] = last
        return arr
```
